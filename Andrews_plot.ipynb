```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib as mpt
from matplotlib import pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
def ft(X,t,i) :
    temp = X[i][0]*np.sqrt(2)+X[i][1]*np.sin(2*np.pi*t)+X[i][2]*np.cos(2*np.pi*t)+X[i][3]*np.sin(3*np.pi*t)
    return temp
```

```python
i=1
t=np.arange(0,1.01,0.01)
while i < 150 :
    col= "C{0}".format(Y[i])
    plt.plot(t,ft(X,t,i),color=col)
    i+=1
plt.show()
```
